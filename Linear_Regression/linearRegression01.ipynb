```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/drive/MyDrive/Machine Learning from scratch/Linear Regression-01/homeprices.csv")
df.head()
```

```python
plt.xlabel('area(sq.ft)')
plt.ylabel('price(USD)')
plt.scatter(df.area,df.price, color='red',marker='+')
```

##training and predicting


```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
plt.xlabel('area(sq.ft)')
plt.ylabel('price(USD)')
plt.scatter(df.area,df.price, color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```

##Getting the slope and intercept

```python
 reg.coef_ #slope
```

```python
reg.intercept_ #value of b
```

##predicting more values

```python
df2 = pd.read_csv("/content/drive/MyDrive/Machine Learning from scratch/Linear Regression-01/areas.csv")
df2.shape
```

```python
pred = reg.predict(df2)
```

```python
df2['price'] = pred
df2.to_csv("/content/drive/MyDrive/Machine Learning from scratch/Linear Regression-01/predicted.csv")
```
